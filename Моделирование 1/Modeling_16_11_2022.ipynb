# Моделирование физических систем №1
## “Разделение изотопов"
**Постановка задачи:** Рассмотрим задачу пространственного разделения ядер с одинаковыми зарядами, но разными массами (изотопов). Из источника О (центра координат) вылетают частицы в направлении оси У перпендикулярно магнитному полю. Приложенное магнитное поле B G параллельно оси z. Изотопы имеют одинаковые по направлению и величине начальные скорости. Необходимо рассчитать траекторию движения атомов с одинаковыми зарядами, но разными массами во внешнем магнитном поле. 

Провести вычислительные эксперименты разделения изотопов при следующих параметрах: масса изотопов лежат в переделах m=(20-24) а.е.м. Вдоль оси Z приложено магнитное поле B=0.01 Тл. Заряд изотопов равна q = 1.6 * 10^(–19) Кл. Скорость вылета частиц из источника О υ = 1000м/с.


### Выполнение работы:
#### Теоретические выкладки:

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.animation as animation
import time
```

#### Обозначим входные параметры программы:

```python
aem = 1.66*pow(10,-27) #1 а.е.м. = 1.66*10^(-27) кг
m_arr = [20*aem, 21*aem, 22*aem, 23*aem, 24*aem] #Массы изотопов в кг. (масса в условии задана в а.е.м.)
B = 0.01 #Индукция магнитного поля в Тл
q = 1.6*pow(10, -19) #Заряд изотопов в Кл
v = 10000 #Скорость вылета частицы из начала координат в м/с
```

#### Построим графики движения изотопов разных масс в магнитном поле, согласно полученным в теоретических выкладках формулам для координат:

```python
m_max = m_arr[4]
t_arr = np.linspace(0, 2*np.pi*m_max/q/B, num = 1000, endpoint=True)

def x(t, A):
    return (-v * np.cos(A*t) / A + v / A)

def y(t, A):
    return (v * np.sin(A*t) / A)

def make_x(ind):
    A = q*B/m_arr[ind]
    x_arr = [ x(t, A) for t in t_arr ]
    return x_arr

def make_y(ind):
    A = q*B/m_arr[ind]
    y_arr = [ y(t, A) for t in t_arr ]
    return y_arr
    

fig = plt.figure(figsize=(13,10))
axs = fig.subplots(1,1)
axs.clear() 
axs.minorticks_on()
axs.grid(True, which='both')
axs.grid(which='minor', color='gray', linestyle=':')

axs.plot(make_x(0), make_y(0), color='grey', label="m = 20 a.e.м.")
axs.plot(make_x(1), make_y(1), color='red', label="m = 21 a.e.м.")
axs.plot(make_x(2), make_y(2), color='blue', label="m = 22 a.e.м.")
axs.plot(make_x(3), make_y(3), color='green', label="m = 23 a.e.м.")
axs.plot(make_x(4), make_y(4), color='purple', label="m = 24 a.e.м.")

axs.set_ylabel('y, м')
axs.set_xlabel(f'x, м \n B = {B} Тл, R = {q*pow(10,19)}' + r' $ \cdot 10^{-19} $ Кл,' + f' v = {v} м/с \n Магнитное поле направлено на нас');


```

На основе полученного графика траектории движения изотопа в магнитном поле можно сделать следующие выводы:

* Заряженная частица в магнитном поле движется по эллиптической орбите за счёт действующей на неё силы Лоренца.
* Чем больше масса частицы, тем больше радиус её траектории.
* Координаты x и y изотопа меняются по косинусоедальному и синусоедальному законам соответственно.
